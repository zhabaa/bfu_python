```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
t = np.linspace(0, 2*np.pi, 1000)
aspect = [(3, 2), (3, 4), (5, 4), (5, 6)]
phi = np.pi / 4
```

```python
plt.figure(figsize=(12, 10))

gs = GridSpec(2, 2, figure=plt.gcf())

for i, (a, b) in enumerate(aspect):
    x = np.sin(a * t + phi)
    y = np.sin(b * t)

    ax = plt.subplot(gs[i])

    ax.plot(x, y, color=f'C{i}')
    ax.set_title(f'{a}:{b}', fontsize=11)

    ax.grid(True)

    ax.set_xlim(-1.3, 1.3)
    ax.set_ylim(-1.3, 1.3)

    ax.text(0.05, 0.95,
            f'x = sin({a}t + π/4)\n' \
            f'y = sin({b}t)',
            transform=ax.transAxes, va='top', ha='left',
            bbox=dict(facecolor='white', alpha=0.5)
    )

plt.suptitle('Фигуры Лиссажу с разными соотношениями частот', fontsize=13)
plt.tight_layout()
plt.show()
```
