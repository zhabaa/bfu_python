```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre
```

```python
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

x = np.linspace(-1, 1, 400)
```

```python
plt.figure(figsize=(10, 6))
plt.title('Полиномы Лежандра', fontsize=13)

plt.xlabel('x', fontsize=11)
plt.ylabel('P|n(x)', fontsize=11)

plt.grid(True)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.xlim(-1, 1)
plt.ylim(-1, 1.1)

for n in range(1, 8):
    Pn = legendre(n)
    y = Pn(x)
    plt.plot(x, y, color=colors[n-1], label=f'n = {n}')

    x_note = 0.8
    y_note = Pn(x_note)

    plt.annotate(f'n = {n}', 
                 xy=(x_note, y_note), 
                 xytext=(x_note + 0.1, y_note),
                 arrowprops=dict(
                     facecolor=colors[n - 1],
                     ),
                 color=colors[n - 1])
```
