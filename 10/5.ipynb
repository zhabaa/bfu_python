```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
def mse(a, b):
    return (a - 1) ** 2 + (b - 2) ** 2
```

```python
a = np.linspace(-3, 5, 100)
b = np.linspace(-1, 5, 100)

A, B = np.meshgrid(a, b)
Z = mse(A, B)

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(A, B, Z, cmap=cm.viridis, alpha=0.8)

ax1.set_title('MSE   ')
ax1.set_xlabel(' a')
ax1.set_ylabel(' b')
ax1.set_zlabel('MSE')

ax2 = fig.add_subplot(122, projection='3d')
surf2 = ax2.plot_surface(A, B, np.log10(Z + 1e-10), cmap=cm.plasma, alpha=0.8)

ax2.set_title('MSE    (log10)')
ax2.set_xlabel(' a')
ax2.set_ylabel(' b')
ax2.set_zlabel('log10(MSE)')

# fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=5)
# fig.colorbar(surf2, ax=ax2, shrink=0.5, aspect=5)

plt.tight_layout()
plt.show()
```
